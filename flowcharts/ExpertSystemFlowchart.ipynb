```python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches

# Cria a figura e os eixos com as configuraes atualizadas
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
fig.patch.set_facecolor('#221913')

# Elementos do fluxograma de um sistema especialista
elements = ["Entrada\nde Dados", "Consulta \nBase de\nConhecimento", 
            "Motor de\nInferncia", "Gerao de\nHipteses,\nSolues", 
            "Interface\nde Usurio", "Feedback\ndo Usurio", 
            "Sada ou\nAo Final"]

# Posies dos elementos em um layout que inclui iteraes
positions = [(0.1, 0.7), (0.3, 0.7), (0.5, 0.7), (0.7, 0.7), 
             (0.7, 0.3), (0.5, 0.3), (0.3, 0.3)]

# Ajuste no tamanho e no preenchimento das caixas
box_padding = 0.02
box_width = 0.10
box_height = 0.10
arrow_padding = 0.02
element_color = "#BD622A"

# Cria retngulos (caixas) e texto
for pos, element in zip(positions, elements):
    ax.add_patch(mpatches.FancyBboxPatch((pos[0] - box_width/2, pos[1] - box_height/2),
                                         box_width, box_height, boxstyle=f"round,pad={box_padding}",
                                         color=element_color, alpha=0.8))
    ax.text(pos[0], pos[1], element, ha='center', va='center', fontsize=10, color='white')

# Exemplo para desenhar as setas entre os elementos do fluxograma

# De 'Entrada de Dados' para 'Consulta  Base de Conhecimento'
ax.annotate('', xy=(positions[1][0] - box_width/2 - arrow_padding, positions[1][1]), 
            xytext=(positions[0][0] + box_width/2 + arrow_padding, positions[0][1]),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# De 'Consulta  Base de Conhecimento' para 'Motor de Inferncia'
ax.annotate('', xy=(positions[2][0] - box_width/2 - arrow_padding, positions[2][1]), 
            xytext=(positions[1][0] + box_width/2 + arrow_padding, positions[1][1]),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# De 'Motor de Inferncia' para 'Gerao de Hipteses/Solues'
ax.annotate('', xy=(positions[3][0] - box_width/2 - arrow_padding, positions[3][1]), 
            xytext=(positions[2][0] + box_width/2 + arrow_padding, positions[2][1]),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# De 'Gerao de Hipteses/Solues' para 'Interface de Usurio'
ax.annotate('', xy=(positions[3][0], positions[4][1] + box_height/2 + arrow_padding * 2), 
            xytext=(positions[3][0], positions[3][1] - box_height/2 - arrow_padding * 2),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# De 'Feedback do Usurio' para 'Interface de Usurio'
ax.annotate('', xy=(positions[5][0] + box_width/2 + arrow_padding, positions[5][1]), 
            xytext=(positions[4][0] - box_width/2 - arrow_padding, positions[5][1]),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# De 'Feedback do Usurio' para 'Sada ou Ao Final'
ax.annotate('', xy=(positions[6][0] + box_width/2 + arrow_padding, positions[6][1]), 
            xytext=(positions[5][0] - box_width/2 - arrow_padding, positions[5][1]),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# Ajuste do arrow padding para setas diagonais
diagonal_arrow_padding = 0.05  # Ajuste conforme necessrio

# Ajuste para a seta de 'Feedback do Usurio' para 'Consulta  Base de Conhecimento'
ax.annotate('', xy=(positions[1][0] + diagonal_arrow_padding, positions[1][1] - box_height/2 - diagonal_arrow_padding), 
            xytext=(positions[5][0] - diagonal_arrow_padding, positions[5][1] + box_height/2 + diagonal_arrow_padding),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5, linestyle='dashed'))

# De 'Feedback do Usurio' para 'Motor de Inferncia' com seta pontilhada
ax.annotate('', xy=(positions[2][0], positions[2][1] - box_height/2 - diagonal_arrow_padding), 
            xytext=(positions[5][0], positions[5][1] + box_height/2 + diagonal_arrow_padding),
            arrowprops=dict(arrowstyle="->", color='white', lw=1.5))


# Configuraes do plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Exibe o fluxograma
plt.show()
```
