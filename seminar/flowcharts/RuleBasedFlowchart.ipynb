```python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches

# Cria a figura e os eixos com as configurações atualizadas
fig, ax = plt.subplots(figsize=(10, 4), dpi=300)  # Ajuste no tamanho para acomodar layout horizontal
fig.patch.set_facecolor('#221913')  # Define a cor de fundo para preto

# Elementos do fluxograma
elements = ["Entrada\nde Dados", "Análise\nde Regras", "Correspondência\nde Regras", "Ação Baseada\nem Regras", "Saída"]

# Posições dos elementos alinhados horizontalmente
positions = [(0.1, 0.5), (0.3, 0.5), (0.5, 0.5), (0.7, 0.5), (0.9, 0.5)]

# Ajuste no tamanho e no preenchimento das caixas
box_padding = 0.03
box_width = 0.10  # Largura da caixa ajustada para layout horizontal
box_height = 0.2

element_color = "#BD622A"  # Cor especificada para os elementos

# Cria retângulos (caixas) e texto com cores e tamanhos atualizados
for pos, element in zip(positions, elements):
    ax.add_patch(mpatches.FancyBboxPatch((pos[0] - box_width/2, pos[1] - box_height/2),
                                         box_width, box_height, boxstyle=f"round,pad={box_padding}",
                                         color=element_color, alpha=0.8))
    ax.text(pos[0], pos[1], element, ha='center', va='center', fontsize=10, color='white')

# Ajuste nas posições das setas para evitar sobreposição com as caixas
arrow_padding = 0.02  # Preenchimento adicional para as setas

# Desenha as setas com posições atualizadas
for i in range(len(positions) - 1):
    ax.annotate('', xy=(positions[i + 1][0] - box_width/2 - arrow_padding, positions[i + 1][1]),
                xytext=(positions[i][0] + box_width/2 + arrow_padding, positions[i][1]),
                arrowprops=dict(arrowstyle="->", color='white', lw=1.5))

# Configurações do plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Exibe o fluxograma
plt.show()
```
